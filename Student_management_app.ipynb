# Function to check grades

```python
def grade_score(grade):
  if not 0 <= grade <= 100:
    raise ValueError(f'Invalid grade {grade}. Value should be between 0 and 100')
  g = grade
  if g >=70: g='A'
  elif g >= 60: g='B'
  elif g >= 50: g='C'
  elif g >= 45: g='D'
  elif g >= 40: g='E'
  else: g='F'

  match g:
        case "A": print("Excellent")
        case "B": print("Very good")
        case "C": print("Good")
        case "D": print("Fair")
        case "E": print("Passed")
        case "F": print("Failed")
  return g

grade_score(100)
```

# Take user input

```python
def score_to_point(score):
    score = int(score)
    if score >= 70: return 5
    elif score >= 60: return 4
    elif score >= 50: return 3
    elif score >= 45: return 2
    elif score >= 40: return 1
    else: return 0
def get_cgpa(grades):
  total_units = 8
  total_points = 0
  for gp in grades:
    total_points += score_to_point(gp)
  cgpa = total_points/total_units
  return round(cgpa, 2)

def collect_info(db, id='6'):
  db[id]= {}
  db[id]['grades']= {}
  db[id]['name'] = input('Please enter your full name: ')
  db[id]['matric'] = input('Please enter your Matric No: ')
  try:
    db[id]['age'] = int(input('Please enter your age: '))
  except ValueError:
    raise ValueError('Value is not a number')

  # Get courses
  i=0
  courses = []
  print("You're allowed to register four courses")
  while (i<4):
    c = input('Enter a course: ')
    courses.append(c)
    i+=1

  db[id]['courses'] = courses
  grades= []
  scores= []

  # Get grade points
  for c in courses:
    score = int(input(f'Enter score for {c}: '))
    grade = grade_score(score)
    scores.append(score)
    grades.append(grade)

  # Populate the 'grades' key
  for c, g in zip(courses, grades):
    db[id]['grades'][c] = g

  cgpa = get_cgpa(scores)
  db[id]['cgpa'] = cgpa
  db[id]['is_active'] = True




```
